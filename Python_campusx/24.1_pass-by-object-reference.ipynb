In Python, the concept is neither strictly "pass by value" nor "pass by reference" as traditionally understood in other programming languages. Instead, Python uses a model often referred to as "pass by object reference" or "pass by assignment". Here’s a detailed explanation:

### Pass by Object Reference

When you pass an argument to a function in Python, you are passing a reference to the object. This means the function receives a reference to the same object in memory, but the behavior depends on whether the object is mutable or immutable.

#### Immutable Objects (e.g., int, str, tuple)

For immutable objects, the reference to the object is passed. However, since the object cannot be modified, reassigning the reference inside the function does not affect the original object.

```python
def modify_immutable(x):
    x = 10
    print("Inside function:", x)

a = 5
modify_immutable(a)
print("Outside function:", a)
```

Output:
```
Inside function: 10
Outside function: 5
```

- `a` remains `5` outside the function because integers are immutable. Reassigning `x` inside the function does not affect `a`.

#### Mutable Objects (e.g., list, dict, set)

For mutable objects, the reference to the object is passed, and the object can be modified in place. Changes to the object inside the function will be reflected outside the function.

```python
def modify_mutable(lst):
    lst.append(4)
    print("Inside function:", lst)

my_list = [1, 2, 3]
modify_mutable(my_list)
print("Outside function:", my_list)
```

Output:
```
Inside function: [1, 2, 3, 4]
Outside function: [1, 2, 3, 4]
```

- `my_list` is modified both inside and outside the function because lists are mutable.

### Summary

- **Pass by Object Reference (Pass by Assignment)**: Python passes a reference to the object. The function receives a reference to the same object, not a copy.
- **Immutable Objects**: The object cannot be modified. Reassigning the reference inside the function does not affect the original object.
- **Mutable Objects**: The object can be modified in place. Changes to the object inside the function affect the original object.

### Visual Representation

   ```
   my_list -----> [1, 2, 3]
   lst     -----> [1, 2, 3]

   then

   my_list -----> [1, 2, 3, 4]
   lst     -----> [1, 2, 3, 4]
   ```

3. **After the function returns**:
   - `my_list` still references the modified list `[1, 2, 3, 4]`.

   ```
   my_list -----> [1, 2, 3, 4]
   ```

### Conclusion

- Python's parameter passing can be best described as "pass by object reference".
- The actual effect of passing an object to a function depends on the mutability of the object. Immutable objects remain unchanged outside the function, while mutable objects can be modified.

### How Variable Referencing Works

### Mutable vs. Immutable Objects

1. **Mutable Objects**:
   - **Definition**: Objects whose contents can be changed after they are created (e.g., lists, dictionaries, sets).
   - **Behavior**: If you modify a mutable object (e.g., by appending to a list or updating a dictionary) inside a function, these modifications will affect the original object because both the actual parameter (outside the function) and the formal parameter (inside the function) reference the same object.

   ```python
   def modify_list(lst):

   my_list = [1, 2, 3]
   modify_list(my_list)
   ```

2. **Immutable Objects**:
   - **Definition**: Objects whose contents cannot be changed after they are created (e.g., integers, strings, tuples).
   - **Behavior**: If you attempt to modify an immutable object (e.g., by reassigning an integer or modifying a string) inside a function, you are actually creating a new object.    def modify_number(x):

- **References**: Both the actual parameter (variable outside the function) and the formal parameter (variable inside the function) point to the same object if the object is mutable. If you change the object (e.g., modify a list), the changes are visible outside the function.
- **Object Creation**: If the object is immutable, modifications within the function do not affect the original object. ### Summary

- **Mutable Objects**: Modifications affect the original object. Both the actual and formal parameters reference the same object.
- **Immutable Objects**: Modifications result in the creation of a new object. The original object remains unchanged.

### Immutable Objects

- **Object Modification**: When dealing with immutable objects (e.g., integers, strings, tuples), you cannot modify the object itself. Instead, any operation that appears to modify it creates a new object.

- **New Object Creation**: In this case, when you perform an operation (e.g., incrementing an integer), a new object is created. The reference in the formal parameter (inside the function) is updated to point to this new object, but the original reference (outside the function) remains unchanged.

**Example with Immutable Object (Integer)**:

```python
def modify_number(x):
    x = x + 1  # Creates a new integer object and assigns it to x

num = 5
modify_number(num)
print(num)  # Output: 5
```

- **Mutable Objects**: The original object is modified directly. Both the actual and formal parameters reference the same object.
